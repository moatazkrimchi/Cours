#Modifier les lignes et les colonnes

```
df.columns
```

```
df.columns = [x.upper() for x in df.columns]
df.columns
```

```
df.head(10)
```

```
df.rename(columns={'g1':'Math','g2':'Physique','g3':'SVT'},inplace=True)
```

```
df.columns = [x.lower() for x in df.columns]
```

```
df.loc[2,'age'] = 18
df.head()
```

```
df['guardian'] = df['guardian'].str.upper()
```

```
df.head()
```

```
def double(absences):
  return absences*2
```

```
df['absences'] = df['absences'].apply(double)
```

```
df.head()
```

```
df['sex'] = df['sex'].map({'F':0, 'M':1}) 
```

```
df.head()
```
